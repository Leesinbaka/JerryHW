```python
# 冰淇淋⼩店是⼀種特殊的餐館。 
# [Restaurant類別]
# 1.寫⼀個Restaurant的類別
# 2.定義包含⼀個函式 「 describe_restaurant 」可以顯⽰餐廳名稱、烹飪的⽅式 

# [IceCreamStand類別]
# 1.編寫⼀個名為IceCreamStand的類別，繼承Restaurant 
# 2.添加⼀個名為flavors的屬性（List），存各種⼝味的冰淇淋組成的列表 

# 編寫show_flavors 方法可以印出 （我喜歡的⼝味 + List裡⾯的內容）
# 1.執行這個IceCreamStand

icecream = ['1.Vanilla – 香草','2.Chocolate – 巧克力','3.Honey almond – 蜂蜜杏仁',
        '4.Butter scotch/pecan – 奶油糖果/核桃','5.Sundae – 聖代']
class Restaurant():
    def __init__(self,name,use):
        self.use = use
        self.name = name
    def describe_restaurant(self):
        print('餐廳名稱:',self.name,'烹飪的⽅式:',self.use)
        
class IceCreamStand(Restaurant):
    def __init__(self,name,use,flavor):
        super().__init__(name,use)
        self.flavor = flavor
    def show_flavors(self):
        print(self.name,'menu:')
        for p in self.flavor:
            print(p)
            
r = Restaurant('f world','use english')
r.describe_restaurant()
i = IceCreamStand('iceland','ice maker',icecream)
i.describe_restaurant()
i.show_flavors()
```

```python
class A(object): 
    def go(self): 
        print ("go A go!") 
    def stop(self): 
        print ("stop A stop!") 
    def pause(self): 
        raise Exception("Not Implemented")
class B(A): 
    def go(self): 
        super(B, self).go() 
        print ("go B go!")
class C(A): 
    def go(self): 
        super(C, self).go() 
        print ("go C go!")
    def stop(self): 
        super(C, self).stop()
        print ("stop C stop!") 
class D(B,C): 
    def go(self): 
        super(D, self).go() 
        print ("go D go!")
    def stop(self): 
        super(D, self).stop()
        print ("stop D stop!") 
    def pause(self): 
        print ("wait D wait!") 
class E(B,C):
    pass
a = A() 
b = B() 
c = C() 
d = D() 
e = E()

a.go() 
b.go() 
c.go() 
d.go() 
e.go() 


a.stop() 
b.stop() 
c.stop() 
d.stop() 
e.stop() 


a.pause() 
b.pause() 
c.pause() 
d.pause() 
e.pause()
```
